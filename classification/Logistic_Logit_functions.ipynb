```r
library(ggplot2)
```

```r
# logistic function
X1 = seq(-10,10,by=0.1)
Y1 = exp(X1)/(1 + exp(X1))
```

```r
# logit/log-odds function (Inverse of logistic function
X2 = seq(0, 1, by=0.1)
Y2 = log(X2 / (1 - X2))
```

```r
X3 = seq(-10, 10, by=0.1)
Y3 = X3
```

```r
df1 = data.frame(X1, Y1)
df2 = data.frame(X2, Y2)
df3 = data.frame(X3, Y3)
```

```r
ggplot() + 
geom_line(data=df1, aes(x=X1, y=Y1), color='green') + 
geom_line(data=df2, aes(x=X2, y=Y2), color='red')  +
geom_line(data=df3, aes(x=X3, y=Y3), color='black', alpha=0.2, linetype = "dashed")  +
scale_colour_manual("",  breaks = c("Logistic Function", "Logit Function"), values = c("green", "red")) +
xlab(" ") +
labs(title="Logistic & Logit Functions")
```
