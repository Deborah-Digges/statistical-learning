# Matrix

```r
## No scalar in R. Everything is a vector. A scalar is a vector of length 1
# Matrix
# Args: data, nrows, ncols
# Arranges data in column major order
z = matrix(seq(1,12), 4, 3)
z
```

```r
# the byrow=TRUE option can be used to populate the matrix in order of the rows.
z = matrix(seq(1,4), 2, 2, byrow=TRUE)
z
```

```r
# square root of each element of a matrix
sqrt(z)
```

```r
# Raise each element of a matrix by an index
# Can even use fractional and negative indices
z ^ 2
```

```r
# Third and fourth row; second and third column
z[3:4, 2:3]
```

```r
# Second and third column
z[,2:3]
```

```r
# First column: This is no longer a matrix but a vector
z[,1]
```

```r
## How do we maintain the matrix status
z[,1, drop=FALSE]
```

```r
# Query the rows and columns of a matrix
dim(z)
```

# Generating & Plotting Data

```r
# Runif -generates n random uniforms
x = runif(1000)
plot(x)
hist(x)
```

```r
# Rnorm - generates a vector of n random normal variables
y = rnorm(1000)
plot(y)
hist(y)
```

```r
# Plot x,y
plot(x,y)
```

```r
# Labels and color
plot(x, y, xlab="Random Uniform", ylab="Random Normal", col="blue")
```

```r
# Find the correlation between two sets of numbers
a = rnorm(50)
b = a + rnorm(50, mean=50, sd=0.1)
plot(a, b, xlab="a", ylab="b")
cor(a, b)
```

```r
# Contour: Used to plot 3d data
# 1. A vector of x values - 1st dimension
# 2. A vector of y values - 2nd dimension
# 3. A matrix whose elements correspond to the z value(the third dimension) for each pair of x,y
# coordinates
x = seq(1, 100)
y = x
f = outer(x,y,function (x,y) x ^ 2 + y ^ 2)
contour(x, y, f)
    
# increase the number of levels
contour(x, y, f, nlevels=45)
```

```r
# Heat map - color coded plot whose colors depend on the z values
image(x, y, f, col = heat.colors(25))
```

```r
# persp: create a 3d plot
persp(x, y, f)
```

```r
# Theta and phi control the angles at which the plot is viewed
persp(x, y, f, theta=30, phi=20)
persp(x, y, f, theta=30, phi=70)
persp(x, y, f, theta=30 ,phi=40)
```

# Indexing

```r
## Index: Starts at 1
x = c(1, 2, 3)
x[1]
```

```r
## Slicing: Inclusive of last index
x[2:3]
```

```r
## Remove the second element from the vector and return the subsetted vector 
## Does not remove the element from the original vector
x[-2]
x
```

```r
## Remove elements at a collection of indices
x[-c(2,3)]
x
```

```r
A = matrix(1:16,4,4)
A
```

```r
# Access a specific element
A[2, 3]
```

```r
# 1st and 3rd rows, 2nd and 4th columns
A[c(1,3), c(2,4)]
```

```r
# 1st to 3rd row, 2nd to 4th column
A[1:3, 2:4]
```

```r
# 1st and 2nd row
A[1:2, ]
```

```r
# 1st and 2nd column
A[, 1:2]
```

```r
# R Treats a single row of a matrix as a vector
A[1,]

# To keep it as a matrix, use drop=FALSE
A[1, ,drop=FALSE]
```

```r
# The use of a negative sign - in the index tells R 
# to keep all rows or columns except those indicated in the index.

# Remove the first and the third row
A[-c(1,3), ]
```
