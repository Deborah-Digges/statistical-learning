```r
## Define a vector
x = c(1, 2, 3)
```

```r
## Sequence Generator
y = seq(from=3, length=3, by=3)
```

```r
## Help on a function
?seq
```

```r
## Supports parallel vector operations
x
y
x + y 
x / y
x ^ y
x * y
```

```r
## Index: Starts at 1
x[1]
```

```r
## Slicing: Inclusive of last index
x[2:3]
```

```r
## Remove the second element from the vector and return the subsetted vector 
## Does not remove the element from the original vector
x[-2]
x
```

```r
## Remove elements at a collection of indices
x[-c(2,3)]
x
```

```r
## No scalar in R. Everything is a vector. A scalar is a vector of length 1
# Matrix
# Args: data, nrows, ncols
# Arranges data in column major order
z = matrix(seq(1,12), 4, 3)
z
```

```r
# Third and fourth row; second and third column
z[3:4, 2:3]
```

```r
# Second and third column
z[,2:3]
```

```r
# First column: This is no longer a matrix but a vector
z[,1]
```

```r
## How do we maintain the matrix status
z[,1, drop=FALSE]
```

```r
# Query the rows and columns of a matrix
dim(z)
```

```r
# rm(var) - remove variable from the working directory
rm(x)
```

```r
# Runif -generates n random uniforms
x = runif(1000)
plot(x)
hist(x)
```

```r
# Rnorm - generates n random normals
y = rnorm(1000)
plot(y)
hist(y)
```

```r
# Plot x,y
plot(x,y)
```

```r
# Labels and color
plot(x, y, xlab="Random Uniform", ylab="Random Normal", col="blue")
```

```r
# Layout - panel of plots with 2 rows and 1 column
# This division will stay in place until you reset it with another mf row command
par(mfrow=c(2,1))
plot(x,y)
hist(y)
```
