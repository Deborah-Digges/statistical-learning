```r
# 95% Confidence Interval

confidenceInterval = function(statistic, standard_error, t_critical) {
     c(statistic - standard_error* t_critical, statistic + standard_error * t_critical);
}
```

```r
price_coeff = -0.05448
price_coeff_stand_error = 0.00523
price_conf_int = confidenceInterval(price_coeff, price_coeff_stand_error, 2)

price_conf_int
```
