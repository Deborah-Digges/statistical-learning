## Part 1 - original data set

```r
options(jupyter.plot_mimetypes = 'image/png')
```

```r
set.seed(1)
X = rnorm(100)
```

```r
eps = rnorm(100, mean=0, sd=0.5)
```

```r
Y = -1 + 0.5 * X + eps
```

```r
length(Y)
# Beta0 = Y-intercept = -1
# Beta1 = Slope = 0.5
```

```r
plot(X, Y)
# Weak Positive relationship between X and Y
```

```r
fit = lm(Y~X)
```

```r
summary(fit)
```

```r
plot(X,Y)
abline(fit, col="red")
abline(-1, 0.5, col="blue")
legend("topleft", c("Least squares line", "Population Regression Line"), col=c("red", "blue"), lty=c(1,1))
```

```r
# The estimated values of beta0 and beta1 are quite close to the true values
```

```r
poly_fit = lm(Y~poly(X,2))
```

```r
summary(poly_fit)
```

```r
plot(X, Y)
points(X, fitted(poly_fit), col='red', pch=20)
# The coefficient estimates differ a little more from the true values
# The quadratic term is not signficant so there is no evidence that the quadratic term improves the model fit even though th
# R^2 is slightly higher and the RSE slightly lower.
confint(poly_fit)
```

```r
# a.)
set.seed(1)
X1 = runif(100)
X2 = 0.5 * X1 + rnorm(100)/10
Y = 2 + 2 * X1 + 0.3 * X2 + rnorm(100)
```

```r
# Y = beta0 + beta1 * X1 + Beta2 * X2 + eps
```

```r
# beta0 = 2
# beta1 = 2
# beta2 = 0.3
```

```r
cor(X1, X2)
```

```r
plot(X1, X2)
```

```r
fit = lm(Y~X1 + X2)
```

```r
summary(fit)
```

```r
# beta0^ is close to beta0
# However beta1^ and beta2^ are not close to their true values
# The coefficient estimate beta2^ is not significant
# We can reject the null hypothesis that H0: beta1 = 0 at the 5% significance level
# We cannot reject the null hypothesis that H0: beta2 = 0 as the p-value is not significant at the 5% significant level
```

```r
fit_x1 = lm(Y~X1)
```

```r
summary(fit_x1)
# we can reject the null hypothesis H0: beta1=0
```

```r
fit_x2 = lm(Y~X2)
```

```r
summary(fit_x2)
# We can reject the null hypothesis H0: beta2 = 0
```

```r
# The two variables X1 and X2 are collinear(they have a high correlation)
# It is difficult to separate out the individual effects of X1 and X2 on Y in a model containing both X1 and X2
# Hence, one of the variables in the 2 variable model will not be significant.
# However, in both the single variable models, each of the individual variables are significant in the absence of the other
```

```r
# Suppose there was unfortunately one additional point which was unfortunately mismeasured
X1=c(X1, 0.1)
```

```r
X2=c(X2, 0.8) 
```

```r
Y = 2 + 2 * X1 + 0.3 * X2 + rnorm(101)
```

```r
plot(X2, X1)
```

```r
fit = lm(Y ~ X1 + X2)
```

```r
plot(fit)
```

```r
# the extra point is a point of high leverage for the model Y~X1+X2
```

```r
fit = lm(Y~X1)
plot(X1, Y)
```

```r
plot(fit)
```

```r
# The extra point is neither an outlier or a high leverage point for the model Y~X1
```

```r
fit = lm(Y~X2)
```

```r
plot(X2, Y)
```

```r
plot(fit)
```

```r
# The point is a high leverage point for the model Y~X2
```

```r
library(MASS)
?Boston
```

```r
head(Boston)
```

```r
fitRegressionAndPlot = function(predictor, df) {
    fit = lm(crim~get(predictor), df)
    print(paste("Predictor:", predictor))
    print(summary(fit))
    plot(crim~get(predictor), df, xlab = predictor)
    abline(fit, col='red')
}
```

```r
predictors = c('zn', 'indus', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat', 'medv')

for(predictor in predictors)
    fitRegressionAndPlot(predictor, Boston)
```

```r
fitPolynomialAndPlot = function(predictor, df) {
    fit = lm(crim~get(predictor) + I(get(predictor) ^2) + I(get(predictor)^3), df)
    print(paste("Predictor:", predictor))
    print(summary(fit))
    plot(crim~get(predictor), df, xlab = predictor)
    points(Boston[[predictor]], fitted(fit), col='red', pch=20)
}
```

```r
predictors = c('zn', 'chas', 'nox', 'rm', 'age', 'dis', 'rad', 'tax', 'ptratio', 'black', 'lstat', 'medv')
# indus
for(predictor in predictors)
    fitPolynomialAndPlot(predictor, Boston)
```
